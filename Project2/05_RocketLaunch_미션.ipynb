```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# TODO: 데이터 불러와서 확인하기
Rocket_df = pd.read_csv("RocketLaunchDataCSV.csv")
Rocket_df.head()
```

```python
Rocket_df.isnull().sum()
```

```python
# TODO: 결측치 처리및 문자열데이터 인코딩
```

```python
# 전체가 누락된 컬럼 : 'Hist Ave Max Wind Speed','Hist Ave Visibility','Hist Ave Sea Level Pressure'
# 날씨와 무관한 컬럼 : 'Name','Date',Time (East Coast)','Location','Day Length','Notes', 'Sea Level Pressure'
# 종속변수 : 'Launched?'
```

```python
# 날씨와 무관한 컬럼 제거

Rocket_df.drop(['Name','Date','Time (East Coast)','Location','Day Length','Notes', 'Sea Level Pressure'],axis=1, inplace=True)
Rocket_df
```

```python
# 종속변수 : 'Launched?'
target = Rocket_df['Launched?']
data = Rocket_df.drop('Launched?',axis=1)
```

```python
# 종속변수의 null 처리 : NaN -> 'N'
target = target.fillna("N")
target
```

```python
# 적당한 값으로 결측치 처리
# crewed = Uncrewed로
# Launche? = N
# Wind Direction = Unknown
# Condition = Cloudy
# 기타는 중앙값 또는 최빈값

data['Crewed or Uncrewed'] = data['Crewed or Uncrewed'].fillna("Uncrewed")
data['Wind Direction'] = data['Wind Direction'].fillna("Unknown")
data['Condition'] = data['Condition'].fillna("Cloudy")
for column in data.columns:
    if data[column].dtype == 'object':  # 문자열 데이터의 경우 최빈값 사용
        mode_value = data[column].mode().iloc
        data[column] = data[column].fillna(mode_value)
    else:  # 숫자 데이터의 경우 중앙값 사용
        median_value = data[column].median()
        data[column] = data[column].fillna(median_value)
data
```

```python
Rocket_df.drop(['Hist Ave Max Wind Speed','Hist Ave Visibility','Hist Ave Sea Level Pressure'],axis=1,inplace=True)
Rocket_df
```

```python
# 문자데이터 인코딩

from sklearn.preprocessing import LabelEncoder

columns = ['Crewed or Uncrewed','Wind Direction','Condition']
le = LabelEncoder()

for column in columns:
    data[column] = le.fit_transform(data[column])
data.drop(['Hist Ave Max Wind Speed','Hist Ave Visibility','Hist Ave Sea Level Pressure'], axis=1, inplace=True)
```

```python
# TODO: 데이터 분리하기
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
# TODO: 모델 만들고 성능 평가하기기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

gridcv = GridSearchCV(DecisionTreeClassifier(), 
                      param_grid={'max_depth': range(1, 11)}, 
                      cv=5)
rf = RandomForestClassifier()
lr = LogisticRegression()

gridcv.fit(X_train, y_train)
y_pred_rf = rf.fit(X_train, y_train).predict(X_test)
y_pred_lr = lr.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.metrics import classification_report


out = f'''
compare DT ,RF, LR
DT : {gridcv.best_score_}
RF : {rf.score(X_test, y_test)}
LR : {lr.score(X_test, y_test)}

DT confusion matrix : 
{classification_report(y_test, gridcv.best_estimator_.predict(X_test))}

RF confusion matrix : 
{classification_report(y_test, y_pred_rf)}

LR confusion matrix : 
{classification_report(y_test, y_pred_lr)}
'''
print(out)
```
